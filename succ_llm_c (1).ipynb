```python
# كتابة الكود في ملف Program.cs
%%writefile MyProject/Program.cs
using System;
using System.Collections.Generic;
using System.Threading.Tasks;
using LLama;
using LLama.Common;

class Program
{
    static async Task Main()
    {
        string modelPath = "/content/phi-2.Q4_K_M.gguf"; // تأكد من صحة مسار الملف

        var parameters = new ModelParams(modelPath)
        {
            ContextSize = 4096,
            GpuLayerCount = 5 // يمكنك تعديل هذا حسب إمكانيات جهازك
        };

        using var model = LLamaWeights.LoadFromFile(parameters);
        using var context = model.CreateContext(parameters);
        var executor = new InteractiveExecutor(context);

        var chatHistory = new ChatHistory();
        chatHistory.AddMessage(AuthorRole.System, "Hello, I'm Bob! How can I help you?");


        Console.WriteLine("Chat session started. Type 'exit' to quit.");
        string userInput = Console.ReadLine() ?? "";

        while (userInput != "exit")
        {
            await foreach (var text in executor.InferAsync(userInput, new InferenceParams()
            {
                MaxTokens = 1024,
                AntiPrompts = new List<string> { "User:" }
            }))
            {
                Console.Write(text);
            }

            userInput = Console.ReadLine() ?? "";
        }
    }
}
```
